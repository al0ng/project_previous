```python
import csv
import os
import requests
import random
import time
import json
```

#### 设置请求头，以及一些常量的值
参考https://github.com/hahaha108/meituanAppSpider 

```python
# -*- coding: utf-8 -*-
'''
meituan api spider's global variables settings.
'''

# HTTP Request headers infos setting variables:
headers = [
    {
        'User-Agent': 'AiMeiTuan /Xiaomi-4.4.2-MI 6 -720x1280-240-5.5.4-254-863254010002128-qqcpd',
        'Host': 'api.meituan.com',
        'Connection': 'Keep-Alive'
    },
    {
        'User-Agent': 'AiMeiTuan /HUAWEI-4.4.2-HUAWEI MLA-AL10-720x1280-240-5.5.4-254-863254010002128-qqcpd',
        'Host': 'api.meituan.com',
        'Connection': 'Keep-Alive'
    }
]

# limit setting variable used to limit numbers of restaurants in one request:
limit = 25

# Data storage path and filename(.csv or .txt file) setting variables:
savePath = ''
filename = 'lng_lat.txt'
```

#### 数据爬取

```python
URL = "http://api.meituan.com/group/v4/deal/select/city/55/cate/1?sort=solds&hasGroup=true&mpt_cate1=1&offset={0}&limit={1}"

# if not os.path.exists(savePath):
#     os.makedirs(savePath)

filePath = os.path.join(savePath,filename)
file = open(filePath, 'a')
counter = 0
# for i in range(0, 1):
while(True):
    time.sleep(1)
    url = URL.format(str(counter*limit), limit)
    counter = counter + 1
    #url = URL.format(str(0), limit)
    print(url)
    response = requests.get(url,headers=random.choice(headers))
    info_dict = json.loads(response.text)
    try:
        info_list = info_dict['data']
        if len(info_list) < 1:
            break;
        print("爬取总条数" + str(counter * 25 + len(info_list)))
        for i in range(0, len(info_list)):
            lng = info_list[i]['poi']['lng']
            lat = info_list[i]['poi']['lat']
            #注意！分隔符为空格
            content = str(lng)+' '+str(lat) + '\n'
            #print(content)
            file.write(content)
    except Exception as e:
#         #time.sleep()
        print(str(e))
        continue
print("Down!")
file.close()
```
